```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


```

## Diagramme à barres

Voyons maintenant ce qu'est un diagramme à barres en en créant un. 
Un diagramme à barres peut être créé de la manière suivante 

```python
df = pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])

df.plot.bar()
```

Pour produire un diagramme à barres empilées, faites stacked=Vrai. 

```python
df = pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])
df.plot.bar(stacked=True)
```

Pour obtenir des diagrammes à barres horizontales, utilisez la méthode barh. 

```python
df = pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])

df.plot.barh(stacked=True)
```

## Histogrammes

Les histogrammes peuvent être tracés en utilisant la méthode plot.hist(). N

```python
df = pd.DataFrame({'a':np.random.randn(1000)+1,'b':np.random.randn(1000),'c':
np.random.randn(1000) - 1}, columns=['a', 'b', 'c'])

df.plot.hist(bins=50)
```

## Diagrammes en boîtes Box Plot

Le diagramme en boites peut être dessiné en appelant Series.box.plot() et DataFrame.box.plot(), ou DataFrame.boxplot() pour visualiser la distribution des valeurs dans chaque colonne.

Par exemple, voici un boxplot représentant cinq essais de 10 observations d'une variable aléatoire uniforme sur[0,1].

```python
df = pd.DataFrame(np.random.rand(10, 5), columns=['A', 'B', 'C', 'D', 'E'])

df.plot.box()
```

## Diagramme de surface

Le graphique de surface peut être créé à l'aide des méthodes Series.plot.area() ou DataFrame.plot.area().

```python
df = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd'])
df.plot.area()
```

## Nuage de points

Le nuage de points peut être créé à l'aide des méthodes DataFrame.plot.scatter().

```python
df = pd.DataFrame(np.random.rand(50, 4), columns=['a', 'b', 'c', 'd'])

df.plot.scatter(x='a', y='b')
```

## Diagramme circulaire

Le diagramme circulaire peut être créé à l'aide de la méthode DataFrame.plot.pie().

```python
df = pd.DataFrame(3 * np.random.rand(4), index=['a', 'b', 'c', 'd'], columns=['x'])
df.plot.pie(subplots=True)
```

## Lignes



```python
from pylab import *
t = arange(0.0, 2.0, 0.01)
s = sin(2.5*pi*t)
plot(t, s)
 
xlabel('Temps (s)')
ylabel('voltage (mV)')
title('Sinusoïde')
grid(True)
show()
```

## Multi graphes

```python

np.random.seed(19680801)
data = np.random.randn(2, 100)

fig, axs = plt.subplots(2, 2, figsize=(5, 5))
axs[0, 0].hist(data[0])
axs[1, 0].scatter(data[0], data[1])
axs[0, 1].plot(data[0], data[1])
axs[1, 1].hist2d(data[0], data[1])

plt.show()
```
