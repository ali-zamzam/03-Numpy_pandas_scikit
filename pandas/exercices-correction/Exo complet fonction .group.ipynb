# 4 - Quelle est la moyenne d'âge par emploi

```python
utilisateurs.groupby("emploi").age.mean()
```

# 5 - Quel est le pourcentage d'homme par emploi et les classer dans l'ordre décroissant

```python
# Création d'une nouvelle colonne
def transforme_sexe(x):
    if x == 'M':
        return 1
    else:
        return 0
    
# on va créer une nouvelle colonne
utilisateurs["sexe_num"] = utilisateurs["sexe"].apply(transforme_sexe)

a = utilisateurs.groupby("emploi").sexe_num.sum() / utilisateurs.emploi.value_counts()*100

a.sort_values(ascending = False)
```

# 6 - Pour chaque emploi, calculer l'âge maximum et minimum

```python
utilisateurs.groupby('emploi').age.agg(['min','max'])
```

# 7 - Pour chaque combinaison emploi-sexe, calculer la moyenne d'âge

```python
utilisateurs.groupby(['emploi','sexe']).age.mean()
```

# 8 - Pour chaque emploi donner la répartition homme - femme

```python
# pour chaque sexe on fait la somme
sexe_emploi = utilisateurs.groupby(['emploi','sexe']).agg('sexe')
sexe_emploi
```

```python
# pour chaque emploi on fait la somme
emploi_nb = utilisateurs.groupby(['emploi']).agg('count')
emploi_nb
```

```python
# je divise
emploi_sexe = sexe_emploi.div(emploi_nb, level='emploi') * 100
emploi_sexe
```
