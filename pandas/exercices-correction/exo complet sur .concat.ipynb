# 1 - Importer pandas et numpy

```python
import pandas as pd
import numpy as np
```

# 2. Créez 3 séries différentes, chacune d'une longueur de 100:

    La première avec des nombres aléatoires de 1 à 4.
    La seconde avec des nombres aléatoires de 1 à 3
    La troisième avec des nombres aléatoires de 10 000 à 30 000.

```python
s1 = pd.Series(np.random.randint(1, high=5, size=100, dtype='int'))
s2 = pd.Series(np.random.randint(1, high=4, size=100, dtype='int'))
s3 = pd.Series(np.random.randint(10000, high=30001, size=100, dtype='int'))
print(s1, s2, s3)

```

# 3 - Créer un DataFrame en joignant les 3 séries par colonne

```python
immobilier = pd.concat([s1, s2, s3], axis=1)
immobilier.head()
```

# 4 - Changer le titre des colonnes en mettant chambre, douche, prixm2

```python
immobilier.rename(columns={0 : 'chambre', 1:'douche', 2:'prixm2'}, inplace=True)
immobilier.head()
```
