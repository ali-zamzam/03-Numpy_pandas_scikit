# 9 - Créer une fonction nommée "majorite"

En fonction de l'argument x, on doit retourner True si la le Nombre d'année est >= à 18


```python
def majorite(x):
    if x >= 18:
        return True
    else:
        return False
        
print(majorite(30))
```
