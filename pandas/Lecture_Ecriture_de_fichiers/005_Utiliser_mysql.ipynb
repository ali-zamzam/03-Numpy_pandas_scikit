```python
import pandas as pd
from sqlalchemy import create_engine 
 
cnx = create_engine('mysql+pymysql://username:password@server:3306/database').connect() 
sql = 'select * from mytable'
df = pd.read_sql(sql, cnx)
```

Le format de la chaîne de connexion est :

mysql+pymysql://nomd'utilisateur:password@host:port/nom_de_la_base_de_données

Format de la chaîne de connexion peut varier suivant les bases de données

## Lire la table entière dans un dataframe en utilisant read_sql_table

```python
df = pd.read_sql_table('clients',cnx)
```

## Lecture uniquement des colonnes sélectionnées

```python
df = pd.read_sql_table('clients', cnx, columns=["nom"])
```
