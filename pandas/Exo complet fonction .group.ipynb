# 4 - Quelle est la moyenne d'âge par emploi

```python
df2 = data.copy()[['emploi','age']].groupby(['emploi']).mean()
df2
```

# 5 - Quel est le pourcentage d'homme par emploi et les classer dans l'ordre décroissant

```python
# Création d'une nouvelle colonne
# on va créer une nouvelle colonne

def transforme(x):
    if x == 'M':
        return 1
    else:
        return 0
data1['sexe_M'] = data1['sexe'].apply(transforme)

df3 = data1.groupby('emploi').sexe_M.sum()/data1.emploi.value_counts()*100

df3


```
