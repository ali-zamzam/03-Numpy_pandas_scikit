```python
import pandas as pd
import numpy as np
```

```python
exam_data  = {'nom': ['Anne', 'Alex', 'catherine', 'Jean', 'Emillie', 'Michel', 'Matieu', 'Laura', 'Kevin', 'Jonas'],
        'note': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
              'moyenne' : [12.0, 9, 16.0, 5.0, 9.0, 20.0, 14.0, 15.0, 8.0, 19.0],
        'qualification': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes'],
         'profession': ['boulanger', 'chanteur', 'boulanger', 'cuisinier', 'conducteur', 'cuisinier', 'livreur', 'pilote de ligne', 'mécanicien', np.NAN]}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df = pd.DataFrame(exam_data , index=labels)
```

Dans ce tutoriel, nous apprendrons comment classer le dataframe dans pandas par ordre croissant et décroissant avec la valeur maximale du rang, la valeur minimale du rang, la valeur moyenne du rang. Nous verrons un exemple pour chacun d'eux.  Nous allons classer les données par rangée en fonction des différentes méthodes.

Dans ce tutoriel, nous traiterons des exemples suivants

    Classer la trame de données par ordre croissant et décroissant.
    Classer le dataframe par rang dense si 2 valeurs trouvées sont identiques.
    Classer la trame de données par rang maximum si les 2 valeurs trouvées sont les mêmes
    Classer la trame de données par rang Minimum si 2 valeurs trouvées sont identiques.
    Classer la trame de données par groupe

## Classement du dataframe dans python pandas sur l'ordre croissant :

Maintenant, classons le dataframe par ordre croissant de score comme indiqué ci-dessous

```python
df['moyenne_rang']=df['moyenne'].rank(ascending=1)
df
df.sort_values('moyenne_rang')
```

## Classement du dataframe dans  l'ordre décroissant :

Classer le dataframe par ordre décroissant de moyenne comme indiqué ci-dessous

```python
df['moyenne_rang']=df['moyenne'].rank(ascending=False)
df
```

## Classer le dataframe par valeur minimale du rang.

Classer par ordre décroissant de score et si deux scores sont identiques, attribuer le rang minimum aux deux scores comme indiqué ci-dessous.

```python
df['moyenne_rang']=df['moyenne'].rank(ascending=False, method='min')
df
```

## Classer le dataframe par valeur maximale du rang.

Classer par ordre décroissant de score et si deux scores sont identiques, attribuer le rang maximum aux deux scores comme indiqué ci-dessous.

```python
df['moyenne_rang']=df['moyenne'].rank(ascending=False, method='max')
df
```

## Classer le dataframe par rang dense.

Classer par ordre décroissant de score et si deux scores sont identiques, attribuer le même rang. Le rang dense ne saute aucun rang (les rangs min et max sont ignorés).

```python
df['moyenne_rang']=df['moyenne'].rank(ascending=False, method='dense')
df
```

## Classer le dataframe dans python pandas par Groupe

Classer les données dans l'ordre décroissant de la moyenne par profession, de sorte que le classement se fait par sujet.

```python
df["groupe_rang"] = df.groupby("profession")["moyenne"].rank(ascending=0,method='dense')
df
```
