```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 10)
```

```python
exam_data  = {'nom': ['Anne', 'Alex', 'catherine', 'Jean', 'Emillie', 'Michel', 'Matieu', 'Laura', 'Kevin', 'Jonas'],
        'note': [12.5, 20, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
               'moyenne': [15, 18, 16.5, 9.1, 11.25, 17.5, 11.0, 13.8, 7.6, 14.3],
        'qualification': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df = pd.DataFrame(exam_data , index=labels)
df
```

```python
# Imaginons que nous voulions définir des catégories en fonction des moyennes
def bon_eleve(row):
    ValMoyenne = row[2]
    ValQualification = row[3]
    
    if (ValMoyenne <10) & (ValQualification=='no'):
        return 'Il faut se bouger'
    elif (ValMoyenne <10) & (ValQualification=='yes'):
        return "Il faut s'améliorer"
    elif (ValMoyenne >10) & (ValQualification=='yes'):
        return "Très bien poursuivre"
    else:
        return "Encore quelques efforts"
```

```python
# pour activer la fonction bon_eleve sur chaque ligne
# on passe axis="columns"

df["classement"] = df.apply(bon_eleve, axis="columns")
```
