Depuis la miss à jour de pandas 0.20  ix est obsolète. Nous verrons uniquement comment utiliser loc, iloc, at, iat, set_value, mais aussi comment accomplir une indexation mixte positionnelle/étiquette.

 ## .loc basée sur l'étiquette
 
Permet de passer des tableaux 1-D en tant qu'indexeurs. Les tableaux peuvent être soit des tranches (sous-ensembles) de l'index ou de la colonne, soit des tableaux booléens de longueur égale à l'index ou aux colonnes.

Note spéciale : lorsqu'un indexeur scalaire est passé, loc peut assigner un nouvel index ou une nouvelle valeur de colonne qui n'existait pas auparavant.

```python
df.loc[df.index[2],'nom']="Alex modifié"
```

```python
df
```

```python
df.loc[df.index[2:5],'note']=1.78
```

```python
df
```

## iloc -basée sur la position

Similaire à localiser sauf avec des positions plutôt qu'avec des valeurs d'index. Toutefois, vous ne pouvez pas affecter de nouvelles colonnes ou de nouveaux index.

```python
df.iloc[2,3]="YES"
```

```python
df
```

```python
df.iloc[2,df.columns.get_loc('qualification')]="NO"
df
```

```python
df.iloc[7:,1:3]=48
df
```

## at - basé sur l'étiquette
Fonctionne très similaire .loc. Ne peut pas fonctionner sur les indexeurs de tableaux. Peut ! assigner de nouveaux index et colonnes.

L'avantage par rapport au localisateur est que c'est plus rapide.
L'inconvénient est que vous ne pouvez pas utiliser les tableaux pour les indexeurs.

```python
df.at["h",'qualification']="SANS"
df
```

```python
df.at[df.index[2], 'qualification'] = 'SANS'
df
```

## iat - basé sur la position
Fonctionne de manière similaire à iloc. Impossible de travailler dans les indexeurs de tableaux. Impossible ! d'attribuer de nouveaux index et de nouvelles colonnes.

L'avantage par rapport à iloc est que c'est plus rapide.
L'inconvénient est que vous ne pouvez pas utiliser les tableaux pour les indexeurs.

```python
df.iat[2,2]=15
df
```

## set_value - basé sur l'étiquette
Fonctionne très similaire au localisateur pour les indexeurs scalaires. Ne peut pas fonctionner sur les indexeurs de tableaux. Peut ! assigner de nouveaux index et colonnes

Avantage Super rapide
Inconvénient Très rapide parce que pandas ne fait pas un tas de contrôles de sécurité. L'utilisation se fait à vos propres risques. De plus, ceci n'est pas destiné à un usage public.



```python
df.set_value(2,1,15)
df
```
