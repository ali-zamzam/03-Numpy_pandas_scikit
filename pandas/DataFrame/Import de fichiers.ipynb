Pour l'instant, nous allons charger des données dans un DataFrame que nous pouvons utiliser pour voir diverses opérations. Ce code n'utilise que six colonnes de données spécifiques dans le fichier.

```python
produits = pd.read_excel("composition-nutritionnelle.xls", index_col=[0],usecols=[6,7,12,13,15,16,17,18])
```
